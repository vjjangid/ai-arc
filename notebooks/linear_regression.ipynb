## Problem Statement

```python
#Let's understand using simple example first

#size is on x plain
x_train = np.array([1.0, 2.0])

#prices on y axis
y_train = np.array([300.0, 500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```
